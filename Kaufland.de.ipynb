```python
table_data=[]
impresse=[]
product_link=[]
title=[]
j=0
p=1

chromedriver_autoinstaller.install()
chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome(options=chrome_options)

for link in links:
    try:     
        if j==50:   
            driver.quit()
            driver = webdriver.Chrome()
            driver.get(link)
            time.sleep(15)
            element1 = driver.find_element(By.XPATH,'//*[@id="onetrust-accept-btn-handler"]')
            element1.click()
            j=0
        driver.get(link)
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser')
        time.sleep(5)
        try:
            element1 = driver.find_element(By.XPATH,'//*[@id="onetrust-accept-btn-handler"]')
            element1.click()
        except:
            pass
        driver.execute_script("window.scrollBy(0, 1600);")
        time.sleep(5)
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser')
        table=soup.find('div',class_='rd-attribute-table')
        if table:
            table_data.append(table.text)
        else:
            table_data.append('N/A')

        try:
            driver.execute_script("window.scrollTo(0, 0);")
            driver.execute_script("window.scrollBy(0, 500);")
            name=driver.find_element(By.XPATH, '//h1[@class="rd-title"]')
            if name:
                title.append(name.text)
            else:
                title.append('N/A')
            try:
                element = driver.find_element(By.XPATH, '//p[@class="rd-info-card__text rd-info-card__text--small rd-info-card__text--gray"]')
                element.click()
            except:
                pass
            time.sleep(5)
            page_source = driver.page_source
            soup = BeautifulSoup(page_source, 'html.parser')
            s=soup.find('div', class_='rd-seller-overlay')
            d=s.find_all('div',class_='rd-accordion')
            if d:
                for n in d:
                    v=n.find('div',class_='rd-accordion__inner-content')
                    if v:
                        impresse.append(v.text)
                        product_link.append(link)

                    else:
                        impresse.append('N/A')
                        product_link.append(link)
                    break
            else:
                impresse.append('N/A')
                product_link.append(link)
        except:
            table_data.pop(-1)
            title.pop(-1)
    #     driver.quit()
        print('Product ',p,' Scrapped.......')
        p+=1 
        j+=1
    except:
        continue
driver.quit()
l = len(impresse)
m = len(table_data)
n = len(title)
o = len(links)
max_length = max(l, m, n, o)
if l < max_length:
    impresse += ['N/A'] * (max_length - l)
if m < max_length:
    table_data += ['N/A'] * (max_length - m)
if n < max_length:
    title += ['N/A'] * (max_length - n)
if o < max_length:
    links += ['N/A'] * (max_length - o)
```
